# Introduction

As an internal calibration, $^{83m}Kr$ proves very useful in a variety of liquid noble detectors. We utilize a solution of $^{83}Rb$ evaporated onto zeolite beads to generate $^{83m}Kr$ with a half-life of $86.2\ days$. From there, $^{83m}Kr$, with a half-life of $1.83\ hr$, generates a $32.1\ keV$ conversion electron, followed by a $9.4\ keV$ conversion electron with a delay half-life of $154.4\ ns$. This decay scheme is shown in Figure 0.

```python
plt.rcParams['figure.figsize'] = (10.0, 10.0)  

plt.imshow(mpimg.imread('../figures/kr83m_decay_scheme.png'))
plt.title('Fig 1: $^{83m}$Kr Decay Scheme')
plt.axis('off')
plt.show()
```

The short half-life of $^{83m}Kr$ and its mother permit an internal calibration without introducing unwanted signals into the experiment. Furthermore, the short half-life between the $32.1\ keV$ and $9.4\ keV$ calibration lines provides a unique signal that is essentially background free.

# The DoubleScatter Treemaker

We now make the steps to identify double scatter events, the signature of the $^{83m}Kr$ calibration. The DoubleScatter treemaker identifies events with two s1s as follows:
* First S1 (s10) attributed to interactions[0]
* Additional interaction required for second S1 (s11)
 * Prioritize 2nd interaction with S1 and S2 distinct from interaction[0]
 * Otherwise select interaction with S1 distinct from interactions[0] but identical S2
* interactions re-ordered by S1 hit_time_mean
* Positions (x,y,z) pulled from interactions (e.g. interactions[0].x)
 
** Note: ** This places a strict requirement of two S1s and at least one S2

```python
# Re-scale z positions for DoubleScatter positions
data['int_0_z_good'] = (data['int_0_drift_time']/self_set_max_drift_time)*-tpc_height
data['int_1_z_good'] = (data['int_1_drift_time']/self_set_max_drift_time)*-tpc_height
```

```python
# Add time and space differences of DoubleScatter events
data['s1_dt_hit_time_mean'] = data['s1_1_hit_time_mean'] - data['s1_0_hit_time_mean']
data['s1_dt_center_time'] = data['s1_1_center_time'] - data['s1_0_center_time']
data['int_dx'] = np.sqrt( (data['int_0_x'] - data['int_1_x'])**2
                         +(data['int_0_y'] - data['int_1_y'])**2
                         +(data['int_0_z_good'] - data['int_1_z_good'])**2 )
data['int_dxy'] = np.sqrt( (data['int_0_x'] - data['int_1_x'])**2
                         +(data['int_0_y'] - data['int_1_y'])**2 )
data['int_dz'] = np.abs( data['int_0_z_good'] - data['int_1_z_good'] )

# Add S1_ratio
data['s1_ratio'] = data['s1_1']/data['s1_0']
```

```python
# DoubleScatter selection
data_ds = data[ data['s1_0_hit_time_mean']>0 ] # removes NAN's
print('We found %i DoubleScatter events out of %i total events \n DoubleScatter acceptance = %f' 
      % (len(data_ds),len(data),len(data_ds)/len(data)))
```

## Afterpulses
We now turn our attention to what seem to be afterpulses. From a visual inspection of Figure 9 we can select $^{83m}Kr$ events by requiring a S1 time difference in roughly $[600\ ns,1000\ ns]$. If we can isolate and exclude afterpulse events we could potentially push the upper bound of this range to roughly $1500\ ns$ or even $2000\ ns$ before background swamps the tail.

### Remarks about Afterpulses
- Should show up as single scatter event, i.e., afterpulse S1 paired with unrelated S2
- Xe afterpulses specifically should be isolated to leaky PMTs

Let's consider events in the S1_dt time window of $[2600\ ns,3400\ ns]$, possibly due to $Xe$ afterpulses; contamination of these events by $^{83m}Kr$ should be minimal this high in S1_dt. Figure 11 A shows the $S1_dt$ histogram with fit, and Figure 11 B shows where events within $1\sigma$ of the mean Xe afterpulse $S1_dt$ land within the $S1_0\ vs.\ S1_1$ space together with the historical $S1_Ratio>0.1$ cut.

# Electron Lifetime

With a satisfactory selection of double s1 $^{83m}Kr$ events we now turn our attention to calibration and detector operation, beginning with the electron lifetime. 

```python
def exp_decay(x,p0,p1):
    return np.exp(-x*p0)*p1
```

```python
plt.rcParams['figure.figsize'] = (12.0, 10.0)    

# Plot only selected events - cool feature, white is transparent! 
m2 = Histdd(-d['z_good'].values,
            d['s2'].values,
            axis_names=['z', 'area'],
            bins=[np.linspace(0,100,100),np.linspace(1,2e4,100)],)
m2.plot(cblabel = 'Count', log_scale = True)

mean_areas = m2.average(axis='z')
plt.step(mean_areas.histogram[7:70], mean_areas.bin_centers[7:70], color='aqua', label = 'Area average', linewidth = 3)

## Fit
x = mean_areas.histogram[7:70]
y = mean_areas.bin_centers[7:70]
guess = [0.1,15e3] # p0 height, p1 zero point,  
popt, pcov = curve_fit(exp_decay, x, y, p0=guess)
fit = exp_decay(x, *popt)
perr = np.sqrt(np.diag(pcov))
chi2, ndf = chisquare_ndf(y,fit)

e_lifetime_cm = 1/popt[0] # cm
e_lifetime = (e_lifetime_cm/tpc_height)*self_set_max_drift_time/ units.us # us

# plt.plot(x,exp_decay(x,50,1e5))
plt.plot(x,fit, label = 'Exp decay fit', color = 'red', linewidth = 2)

#Add the fit results 
#s = ('Fit results: \n $\chi^2$/ndf : %.0f / %.0f \n p0 : %1.3f $\pm$ %1.3f \n p1 : %1.3f $\pm$ %1.3f \n' 
#    % (chi2, ndf, popt[0], perr[0], popt[1], perr[1]))
s = ('Fit results: \n $\chi^2$/ndf : %.0f / %.0f \n $\\tau_e$ : %1.1f $\mu s$\n' 
    % (chi2, ndf, e_lifetime))
bbox_props = dict(boxstyle="Round,pad=0.3", fc="lightgrey", ec="grey", lw=2, alpha = 0.5)
plt.text(60, 13000, s,  va="center",  bbox=bbox_props)

# plt.xscale('log')
set_axis('Depth [-cm]','S2 area [pe]','Figure ??: Kr83m Electron lifetime fit')
# save_plot(title)

plt.legend(loc = 'upper right', fontsize = 12)
plt.show()
plt.rcParams['figure.figsize'] = (12.0, 10.0)    # Back to normal size
```
